# Tabla de Datos Cuantitativos Continuos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos (los 500 datos que generamos antes)
np.random.seed(7)
datos = np.random.uniform(35.5, 40.0, 500)
datos = np.round(datos, 2)

# Definir intervalos: por ejemplo, clases de 0.5 grados de 35.5 a 40.0
intervalos = np.arange(35.5, 40.5, 0.5)

# Crear tabla de frecuencias usando pd.cut y value_counts
categorias = pd.cut(datos, bins=intervalos, right=True, include_lowest=True)

# Frecuencia absoluta
fa = categorias.value_counts().sort_index()

# Frecuencia relativa
fr = fa / fa.sum()

# Frecuencia acumulada
fac = fa.cumsum()

# Frecuencia relativa acumulada
frac = fr.cumsum()

# Crear DataFrame con toda la información
tabla_frecuencias = pd.DataFrame({
    "Intervalo": fa.index.astype(str),
    "Frecuencia Absoluta": fa.values,
    "Frecuencia Relativa": fr.values,
    "Frecuencia Acumulada": fac.values,
    "Frecuencia Relativa Acumulada": frac.values
})

# Ajustar opciones de pandas para mostrar todas las columnas
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# Imprimir la tabla de frecuencias
print(tabla_frecuencias)

# Visualizaciones
fig, ax1 = plt.subplots(figsize=(12, 6))

# Colores diferentes para cada barra
colores = plt.cm.viridis(np.linspace(0, 1, len(tabla_frecuencias)))

# Gráfico de barras para la frecuencia absoluta con colores diferentes
ax1.bar(tabla_frecuencias["Intervalo"], tabla_frecuencias["Frecuencia Absoluta"], color=colores, label='Frecuencia Absoluta')
ax1.set_xlabel("Intervalo")
ax1.set_ylabel("Frecuencia Absoluta", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title("Frecuencia Absoluta y Acumulada")

# Crear un segundo eje para la frecuencia acumulada
ax2 = ax1.twinx()
ax2.plot(tabla_frecuencias["Intervalo"], tabla_frecuencias["Frecuencia Acumulada"], color='orange', marker='o', label='Frecuencia Acumulada')
ax2.set_ylabel("Frecuencia Acumulada", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Añadir leyendas
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Poligono de Frecuencias

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generar los 500 datos de temperatura
np.random.seed(7)
datos = np.random.uniform(35.5, 40.0, 500)
datos = np.round(datos, 2)

# Definir los intervalos de clase
intervalos = np.arange(35.5, 40.5, 0.5)
categorias = pd.cut(datos, bins=intervalos, right=True, include_lowest=True)

# Calcular frecuencia relativa
frecuencia_absoluta = categorias.value_counts().sort_index()
frecuencia_relativa = frecuencia_absoluta / frecuencia_absoluta.sum()

# Calcular los puntos medios de cada intervalo
puntos_medios = [(interval.left + interval.right) / 2 for interval in frecuencia_relativa.index]

# Graficar el polígono de frecuencia relativa
plt.plot(puntos_medios, frecuencia_relativa.values, marker='o', linestyle='-', color='blue')

plt.title('Polígono de Frecuencia Relativa')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frecuencia Relativa')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Diagrama de Pastel

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Datos
np.random.seed(7)
datos = np.random.uniform(35.5, 40.0, 500)
datos = np.round(datos, 2)

# Crear intervalos
intervalos = np.arange(35.5, 40.5, 0.5)
categorias = pd.cut(datos, bins=intervalos, right=True, include_lowest=True)

# Frecuencia relativa
fa = categorias.value_counts().sort_index()
fr = fa / fa.sum()

# Etiquetas
etiquetas = [str(interval) for interval in fr.index]

# Explode para resaltar el porcentaje más alto
explode = [0.1 if i == fr.values.argmax() else 0 for i in range(len(fr))]

# Colores distintos para cada rebanada
colores = plt.cm.tab20(np.linspace(0, 1, len(fr)))

# Crear diagrama de pastel
plt.figure(figsize=(7, 7))
plt.pie(fr, labels=etiquetas, autopct='%1.1f%%', startangle=90,
        colors=colores, explode=explode, shadow=True)

plt.title('Diagrama de Pastel - Frecuencia Relativa')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Datos generados
np.random.seed(7)
datos = np.random.uniform(35.5, 40.0, 500)
datos = np.round(datos, 2)

# Crear intervalos
intervalos = np.arange(35.5, 40.5, 0.5)
categorias = pd.cut(datos, bins=intervalos, right=True, include_lowest=True)

# Frecuencia acumulada
fa = categorias.value_counts().sort_index()
fac = fa.cumsum()

# Limites superiores de cada intervalo (para el eje x)
limites_superiores = [interval.right for interval in fac.index]

# Graficar ojiva
plt.plot(limites_superiores, fac.values, marker='o', linestyle='-', color='green')

plt.title('Ojiva - Frecuencia Acumulada')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frecuencia Acumulada')
plt.grid(True)
plt.tight_layout()
plt.show()
```
