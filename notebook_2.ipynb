# Tabla de Datos Cuantitativos Continuos


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 2. Generar 500 datos continuos aleatorios entre 0 y 500
np.random.seed(1)
datos_continuos = np.random.uniform(0, 500, 500)

# Mismos intervalos de 50 para agrupar
bins = range(0, 551, 50)
intervalos = pd.cut(datos_continuos, bins=bins, right=False)
frecuencia_abs = intervalos.value_counts().sort_index()
frecuencia_rel = frecuencia_abs / len(datos_continuos)
frecuencia_acum = frecuencia_abs.cumsum()

# Crear tabla
tabla_continuos = pd.DataFrame({
    'Intervalo': frecuencia_abs.index.astype(str),
    'Frecuencia Absoluta': frecuencia_abs.values,
    'Frecuencia Relativa': frecuencia_rel.values,
    'Frecuencia Acumulada': frecuencia_acum.values
})

print("Tabla de Frecuencia - Datos Continuos")
print(tabla_continuos)

```

## Histograma

```python
# Histograma
plt.hist(datos_continuos, bins=bins, edgecolor='black')
plt.title("Histograma - Datos Continuos")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

## Diagrama de Barras Horizontales

```python
# Diagrama de barras horizontal
tabla_continuos.plot.barh(x='Intervalo', y='Frecuencia Absoluta', legend=False)
plt.title("Barras Horizontales - Datos Continuos")
plt.xlabel("Frecuencia")
plt.ylabel("Intervalo")
plt.show()
```

## Poligono de Frecuencias

```python
# Polígono de frecuencia relativa
plt.plot(tabla_continuos['Intervalo'], tabla_continuos['Frecuencia Relativa'], marker='o', linestyle='-')
plt.title("Polígono de Frecuencia Relativa - Datos Continuos")
plt.xticks(rotation=45)
plt.xlabel("Intervalo")
plt.ylabel("Frecuencia Relativa")
plt.grid(True)
plt.show()
```

## Diagrama de Pastel

```python
# Pastel
plt.pie(tabla_continuos['Frecuencia Relativa'], labels=tabla_continuos['Intervalo'], autopct='%1.1f%%')
plt.title("Diagrama de Pastel - Frecuencia Relativa")
plt.show()
```

## Ojiva

```python
# Ojiva
plt.plot(tabla_continuos['Intervalo'], tabla_continuos['Frecuencia Acumulada'], marker='o')
plt.title("Ojiva - Frecuencia Acumulada")
plt.xticks(rotation=45)
plt.xlabel("Intervalo")
plt.ylabel("Frecuencia Acumulada")
plt.grid(True)
plt.show()
```
