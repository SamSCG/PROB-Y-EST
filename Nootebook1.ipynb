# Tabla con datos cualitativos

```python
# Generar 500 datos cualitativos aleatorios
np.random.seed(42)
categorias = ['Rojo', 'Azul', 'Verde', 'Amarillo', 'Negro']
datos_cualitativos = np.random.choice(categorias, 500)

# Calcular frecuencias
frecuencia_abs = pd.Series(datos_cualitativos).value_counts()
frecuencia_rel = frecuencia_abs / len(datos_cualitativos)
frecuencia_acum = frecuencia_abs.cumsum()

# Crear tabla de frecuencia
tabla_cualitativos = pd.DataFrame({
    'Categoría': frecuencia_abs.index,
    'Frecuencia Absoluta': frecuencia_abs.values,
    'Frecuencia Relativa': frecuencia_rel.values,
    'Frecuencia Acumulada': frecuencia_acum.values
})

print("Tabla de Frecuencia - Datos Cualitativos")
print(tabla_cualitativos)
```

# Histograma

- Grafica Frecuencia Absoluta

```python
# Gráfica de barras vertical (frecuencia absoluta)
tabla_cualitativos.plot(kind='bar', x='Categoría', y='Frecuencia Absoluta', legend=False)
plt.title("Barras - Frecuencia Absoluta")
plt.xlabel("Categoría")
plt.ylabel("Frecuencia")
plt.show()
```

# Diagrama de Barra Horizontal


```python
# Gráfica de barras horizontal
tabla_cualitativos.plot.barh(x='Categoría', y='Frecuencia Absoluta', legend=False)
plt.title("Barras Horizontales - Frecuencia Absoluta")
plt.xlabel("Frecuencia")
plt.ylabel("Categoría")
plt.show()
```

# Poligono de Frecuencia

- Grafica de Frecuencia Relativa


```python
# Polígono de frecuencia relativa
plt.plot(tabla_cualitativos['Categoría'], tabla_cualitativos['Frecuencia Relativa'], marker='o', linestyle='-')
plt.title("Polígono de Frecuencia Relativa")
plt.xlabel("Categoría")
plt.ylabel("Frecuencia Relativa")
plt.grid(True)
plt.show()
```

# Diagrama de Pastel 

- Grafica de Frecuencias Relativa



```python
# Diagrama de pastel (frecuencia relativa)
plt.pie(tabla_cualitativos['Frecuencia Relativa'], labels=tabla_cualitativos['Categoría'], autopct='%1.1f%%')
plt.title("Diagrama de Pastel - Frecuencia Relativa")
plt.show()
```

# Ojiva

- Frecuencia acumulada

```python
# Ojiva (frecuencia acumulada)
plt.plot(tabla_cualitativos['Categoría'], tabla_cualitativos['Frecuencia Acumulada'], marker='o')
plt.title("Ojiva - Frecuencia Acumulada")
plt.xlabel("Categoría")
plt.ylabel("Frecuencia Acumulada")
plt.grid(True)
plt.show()
```
