```python
# Generar 500 datos cualitativos aleatorios (frutas)
np.random.seed(42)
categorias = ['Manzana', 'Banana', 'Naranja', 'Fresa', 'Kiwi', 
              'Mango', 'Pera', 'Uva', 'Sandía', 'Piña']
datos_cualitativos = np.random.choice(categorias, 500)

# Calcular frecuencias
frecuencia_abs = pd.Series(datos_cualitativos).value_counts()
frecuencia_rel = frecuencia_abs / len(datos_cualitativos)
frecuencia_acum = frecuencia_abs.cumsum()

# Crear tabla de frecuencia
tabla_cualitativos = pd.DataFrame({
    'Categoría': frecuencia_abs.index,
    'Frecuencia Absoluta': frecuencia_abs.values,
    'Frecuencia Relativa': frecuencia_rel.values,
    'Frecuencia Acumulada': frecuencia_acum.values
})

print("Tabla de Frecuencia - Datos Cualitativos")
print(tabla_cualitativos)

```

# Histograma

- Grafica Frecuencia Absoluta 


```python

# # Gráfica de barras vertical (frecuencia absoluta) con colores diferentes
plt.figure(figsize=(10, 6))
colores_barras = plt.cm.get_cmap('tab10', len(tabla_cualitativos))  # Obtenemos una colormap
tabla_cualitativos.plot(kind='bar', x='Categoría', y='Frecuencia Absoluta', legend=False, color=colores_barras.colors)
plt.title("Barras - Frecuencia Absoluta de Frutas")
plt.xlabel("Fruta")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Diagrama de Barra Horizontal


```python
# Gráfica de barras horizontal
plt.figure(figsize=(10, 6))
tabla_cualitativos.plot.barh(x='Categoría', y='Frecuencia Absoluta', legend=False, color=colores_barras.colors)
plt.title("Barras Horizontales - Frecuencia Absoluta de Frutas")
plt.xlabel("Frecuencia")
plt.ylabel("Fruta")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# Poligono de Frecuencia

- Grafica de Frecuencia Relativa


```python
# Polígono de frecuencia relativa
plt.figure(figsize=(10, 6))
plt.plot(tabla_cualitativos['Categoría'], tabla_cualitativos['Frecuencia Relativa'], marker='o', linestyle='-', color='orange')
plt.title("Polígono de Frecuencia Relativa de Frutas")
plt.xlabel("Fruta")
plt.ylabel("Frecuencia Relativa")
plt.grid(True)
plt.show()
```

# Diagrama de Pastel 

- Grafica de Frecuencias Relativa



```python
# Diagrama de pastel (frecuencia relativa)
plt.figure(figsize=(8, 8))
plt.pie(tabla_cualitativos['Frecuencia Relativa'], labels=tabla_cualitativos['Categoría'], autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700', '#FFB6C1', '#ADD8E6', '#90EE90', '#FFA07A', '#DDA0DD'])
plt.title("Diagrama de Pastel - Frecuencia Relativa de Frutas")
plt.axis('equal')  # Para que el pastel sea circular
plt.show()
```

# Ojiva

- Frecuencia acumulada

```python
# Ojiva (frecuencia acumulada)
plt.figure(figsize=(10, 6))
plt.plot(tabla_cualitativos['Categoría'], tabla_cualitativos['Frecuencia Acumulada'], marker='o', color='purple')
plt.title("Ojiva - Frecuencia Acumulada de Frutas")
plt.xlabel("Fruta")
plt.ylabel("Frecuencia Acumulada")
plt.grid(True)
plt.show()
```
