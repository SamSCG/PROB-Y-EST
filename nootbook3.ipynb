## Tabla con datos cuantitativos discretos


```python
# 1. Generar 500 datos enteros aleatorios entre 0 y 500
np.random.seed(0)
datos_discretos = np.random.randint(0, 501, 500)

# Agrupar en intervalos de 50 (por ejemplo: 0–49, 50–99, ...)
bins = range(0, 551, 50)
intervalos = pd.cut(datos_discretos, bins=bins, right=False)
frecuencia_abs = intervalos.value_counts().sort_index()
frecuencia_rel = frecuencia_abs / len(datos_discretos)
frecuencia_acum = frecuencia_abs.cumsum()

# Crear tabla
tabla_discretos = pd.DataFrame({
    'Intervalo': frecuencia_abs.index.astype(str),
    'Frecuencia Absoluta': frecuencia_abs.values,
    'Frecuencia Relativa': frecuencia_rel.values,
    'Frecuencia Acumulada': frecuencia_acum.values
})

print("Tabla de Frecuencia - Datos Discretos")
print(tabla_discretos)
```

## Histograma
- Grafica Frecuencia Absoluta

```python
# Histograma
plt.hist(datos_discretos, bins=bins, edgecolor='black')
plt.title("Histograma - Datos Discretos")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

## Diagrama de Barra Horizontales

```python
# Diagrama de barras horizontal
tabla_discretos.plot.barh(x='Intervalo', y='Frecuencia Absoluta', legend=False)
plt.title("Barras Horizontales - Datos Discretos")
plt.xlabel("Frecuencia")
plt.ylabel("Intervalo")
plt.show()
```

## Poligono de Frecuencia
- Grafica de frecuencia relativa

```python
# Polígono de frecuencia relativa
plt.plot(tabla_discretos['Intervalo'], tabla_discretos['Frecuencia Relativa'], marker='o', linestyle='-')
plt.title("Polígono de Frecuencia Relativa - Datos Discretos")
plt.xticks(rotation=45)
plt.xlabel("Intervalo")
plt.ylabel("Frecuencia Relativa")
plt.grid(True)
plt.show()
```

## Diagrama de pastel
- Grafica de frecuencia relativa

```python
# Pastel
plt.pie(tabla_discretos['Frecuencia Relativa'], labels=tabla_discretos['Intervalo'], autopct='%1.1f%%')
plt.title("Diagrama de Pastel - Frecuencia Relativa")
plt.show()
```

## Ojiva
- Frecuencia acumulada

```python
# Ojiva
plt.plot(tabla_discretos['Intervalo'], tabla_discretos['Frecuencia Acumulada'], marker='o')
plt.title("Ojiva - Frecuencia Acumulada")
plt.xticks(rotation=45)
plt.xlabel("Intervalo")
plt.ylabel("Frecuencia Acumulada")
plt.grid(True)
plt.show()
```
